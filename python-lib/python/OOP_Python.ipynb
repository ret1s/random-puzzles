```
"""
If we want to support:
p = Point(x=1.5, y=4, z = 6)
OR
p = Point(1.5, 4, 6)
"""
class Point(object):
  """
  A point in space
  """

  def __init__(self, x, y, z):
    self.x = x; self.y = y; self.z = z

p = Point(1.5, 4, 6)
print("({:<6.2f}, {:^6.2f}, {:>6.2f})".format(p.x, p.y, p.z))
```

## Methods

### Instance-level

```
import math
class Vector(object):
  """
  A vector in space
  """
  def __init__(self, x, y, z):
    self.x = x; self.y = y; self.z = z

  def __str__(self, width=6, prec=2):
    return "({:<{w}.{p}f}, {:^{w}.{p}f}, {:>{w}.{p}f})".format(self.x, self.y, self.z, w=width, p=prec)


  def __repr__(self, width=6, prec=2):
    """
    Representation (description) for instances
    """
    return str(self)

  def length(self):
    """
    Return the length of this vector
    """
    return math.sqrt(self.x**2 + self.y**2 + self.z**2)


  def normalize(self):
    """
    Return normalize this vector and return this vector
    """
    print("normalize")

v = Vector(1,2,3)
print(v)
print(v.length())
v.normalize()
```

```
class Vector(object):
  """
  A vector in space
  """
  SIZE = 12 #pt
  def __init__(self, x, y, z):
    self.x = x; self.y = y; self.z = z

  def __str__(self, width=6, prec=2):
    return "({:<{w}.{p}f}, {:^{w}.{p}f}, {:>{w}.{p}f})".format(self.x, self.y, self.z, w=width, p=prec)


  def __repr__(self, width=6, prec=2):
    """
    Representation (description) for instances
    """
    return str(self)

  def length(self):
    """
    Return the length of this vector
    """
    print("length")
  def normalize(self):
    """
    Return normalize this vector and return this vector
    """
    print("normalize")

  @staticmethod
  def from_to(a:Point, b:Point):
    """
    Create and return a vector from point "a" to b "to"
    """
    return Vector(b.x - a.x, b.y - a.y, b.z - a.z)


P = Point(1,2,3)
Q = Point(4,5,6)
PQ = Vector.from_to(P, Q)
print(PQ)
```

## Inherit data + behaviors

```
class Point(object):
  """
  A point in space
  """
  def __init__(self, x, y, z):
    self.x = x; self.y = y; self.z = z

  def __str__(self, width=6, prec=2):
    return "({:<{w}.{p}f}, {:^{w}.{p}f}, {:>{w}.{p}f})".format(self.x, self.y, self.z, w=width, p=prec)


class Shape2D(object):
  def __init__(self, center:Point=Point(0,0,0)):
    self.center = center # the reference point of shapes

  def area(self):
    """
    Compute + return this shape's area
    """
    print("area defined in shape")

  def perimeter(self):
    """
    Compute + return this shape's perimeter
    """
    print("perimeter defined in shape")

class Ellipse(Shape2D):
  pass

class Circle(Ellipse):
  pass

class Polygon(Shape2D):
  pass

class Rectange(Polygon):
  pass


class Square(Rectange):
  pass

s = Square(Point(5,5,5))
s.area()
print(s.center)
```

## Call super constructor

```
class Shape2D(object):
  def __init__(self, center=Point(0,0,0)):
    self.center = center # the reference point of shapes

  def area(self):
    """
    Compute + return this shape's area
    """
    print("area defined in shape")

  def perimeter(self):
    """
    Compute + return this shape's perimeter
    """
    print("perimeter defined in shape")

class Ellipse(Shape2D):
  def __init__(self, center:Point, radius1:float, radius2:float):
    super().__init__(center)
    self.radius1 = radius1
    self.radius2 = radius2

c = Ellipse(Point(1,2,3), 5, 10)
print(c.center)
print(c.radius1)
print(c.radius2)
```

## Call super methods

```
class A:
  def foo(self):
    return 5

class B(A):
  def foo(self):
    return super().foo()**2

obj = B()
print(obj.foo())
```

```
"""
P and Q are points
Short syntax for creating vector PQ: PQ = Q - P
How to?
=> override a suitable specicial functions
"""


class Point(object):
  """
  A point in space
  """
  def __init__(self, x, y, z):
    self.x = x; self.y = y; self.z = z

  def __str__(self, width=6, prec=2):
    return "({:<{w}.{p}f}, {:^{w}.{p}f}, {:>{w}.{p}f})".format(self.x, self.y, self.z, w=width, p=prec)

  def __repr__(self):
      return str(self)

  def __sub__(self, rhs):
    return Vector(self.x - rhs.x, self.y - rhs.y, self.z - rhs.z)


P = Point(1,2,3)
Q = Point(3,4,5)
v = Vector.from_to(P, Q)
PQ = Q - P
v2 = Q.__sub__(P)

print(v)
print(PQ)
print(v2)
```

```
"""
P: point; d: unit Vector
Short syntax for creating point Q distance from P in direction of d with "length of d" units: Q = P + d
How to?
=> override a suitable specicial functions
"""


class Point(object):
  """
  A point in space
  """
  def __init__(self, x, y, z):
    self.x = x; self.y = y; self.z = z

  def __str__(self, width=6, prec=2):
    return "({:<{w}.{p}f}, {:^{w}.{p}f}, {:>{w}.{p}f})".format(self.x, self.y, self.z, w=width, p=prec)

  def __repr__(self):
      return str(self)

  def __sub__(self, rhs):
    return Vector(rhs.x - self.x, rhs.y - self.y, rhs.z - self.z)

  def __add__(self, rhs:Vector):
    return Point(self.x + rhs.x, self.y + rhs.y, self.z + rhs.z)

P = Point(1,2,3)
d = Vector(2,2,2)
Q = P + d
Q2 = P.__add__(d)

print(Q)
print(Q2)
```
